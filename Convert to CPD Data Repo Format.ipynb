```python
from __future__ import absolute_import, division, print_function, unicode_literals

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
complaint_df = pd.read_csv('data/combined_complaints_dedupe.csv', dtype={'ACTION_TAKEN': str})
complaint_df.head()
```

```python
complaint_df['INITIAL_CATEGORY_CODE'] = complaint_df['INITIAL_COMPLAINT_CATEGORY'].apply(lambda x: re.search(r'^(\d{2}\w)', x).group(0))
complaint_df['INITIAL_CATEGORY'] = complaint_df['INITIAL_COMPLAINT_CATEGORY'].apply(lambda x: re.sub(r'^(\d{2}\w)*', '', x).strip())
complaint_df['CURRENT_CATEGORY_CODE'] = complaint_df['FINAL_COMPLAINT_CATEGORY'].apply(lambda x: re.search(r'^(\d{2}\w)', x).group(0) if re.search(r'^(\d{2}\w)', x) else '')
complaint_df['CURRENT_CATEGORY'] = complaint_df['FINAL_COMPLAINT_CATEGORY'].apply(lambda x: re.sub(r'^(\d{2}\w)*', '', x).strip())
complaint_df.head()
```

```python
complaint_df['ACCUSEDOFFICER_FNAME'] = complaint_df['NAME_NO_INITIAL'].apply(lambda x: x.split(',')[1] if len(x.split(',')) > 1 else '')
complaint_df['ACCUSEDOFFICER_LNAME'] = complaint_df['NAME_NO_INITIAL'].apply(lambda x: x.split(',')[0] if len(x.split(',')) > 0 else '')
complaint_df = complaint_df.rename(columns={'CR_NO': 'COMPLAINT_NUMBER', 'FINAL_FINDING': 'FINDING_CODE'})
complaint_df.head()
```

```python
# Some officers only have first initials, pulling those from original column
complaint_df.loc[complaint_df['ACCUSEDOFFICER_FNAME'] == '', 'ACCUSEDOFFICER_FNAME'] = complaint_df.loc[complaint_df['ACCUSEDOFFICER_FNAME'] == '']['NAME'].apply(lambda x: x.split(',')[1])
complaint_df.head()
```

```python
complaint_df.columns.values
```

```python
complaint_df = complaint_df.rename(columns={'ACTION_DESCR': 'ACTION_DESCRIPTION', 'NAME': 'ACCUSEDOFFICER_FULL_NAME',
                                            'UNIT': 'ACCUSEDOFFICER_UNIT', 'DATE_OF_APPOINTMENT': 'ACCUSEDOFFICER_APPOINTMENT_DATE',
                                            'ACTION_TAKEN': 'ACTION_CODE'})
complaint_df['ACCUSEDOFFICER_FNAME'] = complaint_df['ACCUSEDOFFICER_FNAME'].str.strip()
complaint_df['ACCUSEDOFFICER_LNAME'] = complaint_df['ACCUSEDOFFICER_LNAME'].str.strip()
```

```python
complaint_df_clean = complaint_df[['COMPLAINT_NUMBER', 'INCIDENT_DATE', 'COMPLAINT_DATE', 'CLOSED_DATE', 
                                   'ACCUSEDOFFICER_FNAME', 'ACCUSEDOFFICER_LNAME', 'ACCUSEDOFFICER_FULL_NAME', 
                                   'ACCUSEDOFFICER_UNIT', 'ACCUSEDOFFICER_APPOINTMENT_DATE', 'INITIAL_CATEGORY_CODE', 'INITIAL_CATEGORY',
                                   'CURRENT_CATEGORY_CODE', 'CURRENT_CATEGORY', 'ACTION_CODE', 'ACTION_DESCRIPTION', 'FINDING_CODE',
                                   ]].copy()
complaint_df_clean.head()
```

```python
complaint_df_clean.to_csv('data/complaints-cpd-2016-oct.csv',index=False)
```
